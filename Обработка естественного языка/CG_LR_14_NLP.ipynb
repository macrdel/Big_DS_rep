```python
with open(fname, 'r', encoding = "utf_8", errors="ignore") as read_txt:
        txt = read_txt.read()
        txt = re.sub(r'[^А-я #\n]', '', txt)
        txt = txt.lower()
        txt = txt.split('\n')
        for i in range(len(txt)):
            txt[i] = nltk.word_tokenize(txt[i])
            txt[i] = [morph.parse(word)[0].normal_form for word in txt[i] if word not in stop_words]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
k_split = 0.2
x_trn, x_vl, y_trn, y_vl = train_test_split(x, y, test_size = k_split, shuffle = True)
```

```python
vec = CountVectorizer(token_pattern = '\w+', binary = False)
len_trn = len(x_trn)
x_trn = np.append(x_trn, x_vl, axis = 0) # Объединяем x_trn и x_vl и получаем полный корпус
x_trn = vec.fit_transform(x_trn)
x_vl = x_trn[len_trn:]
x_trn = x_trn[:len_trn]
# Преобразуем разреженные матрицы в массивы
x_trn = np.float32(x_trn.toarray())
x_vl = np.float32(x_vl.toarray())
```

```python
import time
```

```python
def train(classifier):
    start = time.time()
    classifier.fit(x_trn, y_trn)
    print('Время вычислений:', time.time() - start)
    print('Оценка точности классификации')
    score1 = classifier.score(x_vl, y_vl)
    print('Точность на проверочном множестве:', round(score1, 4))
    score2 = classifier.score(x_trn, y_trn)
    print('Точность на обучающем множестве:', round(score2, 4))
    return classifier, score1, score2
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
classifier1 = KNeighborsClassifier(n_neighbors=12)
classifier1, score11, score12 = train(classifier1)
```

```python
classifier2 = QuadraticDiscriminantAnalysis()
classifier2, score21, score22 = train(classifier2)
```

```python
classifier3 = LinearDiscriminantAnalysis()
classifier3, score31, score32 = train(classifier3)
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
classifier4 = PassiveAggressiveClassifier()
```

```python
classifier4, score41, score42 = train(classifier4)
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred1 = classifier1.predict(x_vl)
```
