```python
import numpy as np
from keras.models import Model
from keras.layers import Dense, Input
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-1, 1, 300)
np.random.shuffle(X)   
Y = 0.5 * X + 2 + np.random.normal(0, 0.05, (300, ))
plt.scatter(X, Y, marker = 'o', c = 'blue', edgecolor = 'black')
plt.grid()
plt.show()
```

```python
x_train, y_train = X[:250], Y[:250]    
x_test, y_test = X[250:], Y[250:]      
```

```python
inp = Input(shape = (1, ))
output = Dense(units = 1, input_dim = 1)(inp)
model = Model(inp, output)
model.compile(loss='mse', optimizer='sgd')
```

```python
model.fit(x_train, y_train, verbose = 2, epochs = 500, batch_size = 8, validation_data = (x_test, y_test))
```

```python
cost = model.evaluate(x_test, y_test, batch_size= 40)
print('test cost:', cost)
```

```python
a, b = model.layers[1].get_weights()
print(a[0, 0], '\t', b[0])
```

```python
plt.plot(X, a[0, 0] * X + b[0])
plt.scatter(X[:100], Y[:100], marker = 'o', c = 'blue', edgecolor = 'black')
plt.grid()
plt.show()
```
